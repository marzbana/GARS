```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#staable code
pass_rates_stable_syncode = [16.84, 36.29, 46.92, 56.69, 63.16]  # in percentages
pass_rates_stable_syncode_v = np.std(pass_rates_stable_syncode)
# Data
pass_rates_stable = [18.80, 42.09, 52.98, 62.72, 80.00]  # in percentages
pass_rates_stable_v = np.std(pass_rates_stable)
k_values = [1, 5, 10, 20, 50]
gflops_per_pass = 682.71

# Compute TFLOPS
tflops = [k * gflops_per_pass / 1000 for k in k_values]  # Convert GFLOPS to TFLOPS

# Plotting both datasets
plt.figure(figsize=(8, 6))
plt.plot(tflops, pass_rates_stable_syncode, marker='o', linestyle='-', label=f"Syncode (STD: {pass_rates_stable_syncode_v:.2f})")
plt.plot(tflops, pass_rates_stable, marker='s', linestyle='--', label=f"No Syncode (STD: {pass_rates_stable_v:.2f})")
plt.xlabel("TFLOPS", fontsize=12)
plt.ylabel("Passing Rate (%)", fontsize=12)
plt.title("TFLOPS vs Passing Rate (Comparison): Stable-Code-3b", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
#llama instruct
pass_rates_llama_syncode = [2.4, 11.05, 19.98,32.85,50.00]  # in percentages
pass_rates_llama_syncode_v = np.std(pass_rates_llama_syncode)

# Data
pass_rates_llama = [4.32, 16.74,26.39,38.71,57.89]  # in percentages
pass_rates_llama_v = np.std(pass_rates_llama)
k_values = [1, 5, 10, 20, 50]
tflops_per_pass = 10.15

# Compute TFLOPS
tflops = [k * tflops_per_pass for k in k_values]  # Convert GFLOPS to TFLOPS

# Plotting both datasets
plt.figure(figsize=(8, 6))
plt.plot(tflops, pass_rates_llama_syncode, marker='o', linestyle='-', label=f"Syncode (STD: {pass_rates_llama_syncode_v:.2f})")
plt.plot(tflops, pass_rates_llama, marker='s', linestyle='--', label=f"No Syncode (STD: {pass_rates_llama_v:.2f})")
plt.xlabel("TFLOPS", fontsize=12)
plt.ylabel("Passing Rate (%)", fontsize=12)
plt.title("TFLOPS vs Passing Rate (Comparison): CodeLlama-7b-Instruct", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
#llama instruct
pass_rates_open_syncode = [50.67, 77.93, 86.0,91.18, 94.44]  # in percentages
pass_rates_open_syncode_v = np.std(pass_rates_open_syncode)

#measure variance
# Data
pass_rates_open = [68.24, 89.92,93.40,94.10,94.12]  # in percentages
pass_rates_open_v = np.std(pass_rates_open)
k_values = [1, 5, 10, 20, 50]
tflops_per_pass = 1.69

# Compute TFLOPS
tflops = [k * tflops_per_pass for k in k_values]  # Convert GFLOPS to TFLOPS

# Plotting both datasets
plt.figure(figsize=(8, 6))
plt.plot(tflops, pass_rates_llama_syncode, marker='o', linestyle='-', label=f"Syncode (STD: {pass_rates_open_syncode_v:.2f})")
plt.plot(tflops, pass_rates_llama, marker='s', linestyle='--', label=f"No Syncode (STD: {pass_rates_open_v:.2f})")
plt.xlabel("TFLOPS", fontsize=12)
plt.ylabel("Passing Rate (%)", fontsize=12)
plt.title("TFLOPS vs Passing Rate (Comparison): OpenCodeInterpreter-DS-6.7B", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Stable Code
pass_rates_stable_syncode = [16.84, 36.29, 46.92, 56.69, 63.16]
pass_rates_stable_syncode_v = np.std(pass_rates_stable_syncode)
pass_rates_stable = [18.80, 42.09, 52.98, 62.72, 80.00]
pass_rates_stable_v = np.std(pass_rates_stable)
gflops_per_pass_stable = 682.71
tflops_stable = [k * gflops_per_pass_stable / 1000 for k in [1, 5, 10, 20, 50]]

# Llama Instruct
pass_rates_llama_syncode = [2.4, 11.05, 19.98, 32.85, 50.00]
pass_rates_llama_syncode_v = np.std(pass_rates_llama_syncode)
pass_rates_llama = [4.32, 16.74, 26.39, 38.71, 57.89]
pass_rates_llama_v = np.std(pass_rates_llama)
tflops_per_pass_llama = 10.15
tflops_llama = [k * tflops_per_pass_llama for k in [1, 5, 10, 20, 50]]

# OpenCode Interpreter
pass_rates_open_syncode = [50.67, 77.93, 86.0, 91.18, 94.44]
pass_rates_open_syncode_v = np.std(pass_rates_open_syncode)
pass_rates_open = [68.24, 89.92, 93.40, 94.10, 94.12]
pass_rates_open_v = np.std(pass_rates_open)
tflops_per_pass_open = 1.69
tflops_open = [k * tflops_per_pass_open for k in [1, 5, 10, 20, 50]]

# Create a figure with three subplots side by side
fig, axs = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Stable Code Plot
axs[0].plot(tflops_stable, pass_rates_stable_syncode, marker='o', linestyle='-', label=f"Syncode (STD: {pass_rates_stable_syncode_v:.2f})")
axs[0].plot(tflops_stable, pass_rates_stable, marker='s', linestyle='--', label=f"No Syncode (STD: {pass_rates_stable_v:.2f})")
axs[0].set_title("Stable-Code-3b", fontsize=14)
axs[0].set_xlabel("TFLOPS", fontsize=12)
axs[0].set_ylabel("Passing Rate (%)", fontsize=12)
axs[0].grid(True, linestyle='--', alpha=0.7)
axs[0].legend()

# Llama Instruct Plot
axs[1].plot(tflops_llama, pass_rates_llama_syncode, marker='o', linestyle='-', label=f"Syncode (STD: {pass_rates_llama_syncode_v:.2f})")
axs[1].plot(tflops_llama, pass_rates_llama, marker='s', linestyle='--', label=f"No Syncode (STD: {pass_rates_llama_v:.2f})")
axs[1].set_title("CodeLlama-7b-Instruct", fontsize=14)
axs[1].set_xlabel("TFLOPS", fontsize=12)
axs[1].grid(True, linestyle='--', alpha=0.7)
axs[1].legend()

# OpenCode Interpreter Plot
axs[2].plot(tflops_open, pass_rates_open_syncode, marker='o', linestyle='-', label=f"Syncode (STD: {pass_rates_open_syncode_v:.2f})")
axs[2].plot(tflops_open, pass_rates_open, marker='s', linestyle='--', label=f"No Syncode (STD: {pass_rates_open_v:.2f})")
axs[2].set_title("OpenCodeInterpreter-DS-6.7B", fontsize=14)
axs[2].set_xlabel("TFLOPS", fontsize=12)
axs[2].grid(True, linestyle='--', alpha=0.7)
axs[2].legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
